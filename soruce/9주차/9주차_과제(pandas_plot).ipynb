코로나19 확잔자 수를 담고 있는 pandas Series 데이터를 가로막대형 차트로 작성하기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

sys_font = fm.findSystemFonts()
nanum_font = [f for f in sys_font if 'Nanum' in f]
nanum_font
```

```python
fontPath = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['font.size'] = 10
fm._rebuild()

covid_19 = pd.Series([35628, 9176, 31946, 3843, 5488, 2296, 2886, 4670],
                     index=['서울', '대구', '경기', '경북', '인천', '광주', '충남', '부산'])
plt.barh(covid_19.index, covid_19, label='확진자수', color='orange')
plt.legend()
plt.title('2021년 04월 19일 COVID19 확진현황')
plt.show()
```

가로막대형 차트에 값 레이블 표시하기

```python
covid_19 = pd.Series([35628, 9176, 31946, 3843, 5488, 2296, 2886, 4670],
                     index=['서울', '대구', '경기', '경북', '인천', '광주', '충남', '부산'])
plt.barh(covid_19.index, covid_19, label='확진자수', color='orange')
plt.legend()
plt.title('2021년 04월 19일 COVID19 확진현황')

for i, v in enumerate(covid_19.index):
  plt.text(covid_19[i], v, covid_19[i],
           fontsize = 10,
           color='blue',
           horizontalalignment='left',
           verticalalignment='center'
           )
  plt.show()
```

```python
covid_19 = pd.Series([35628, 9176, 31946, 3843, 5488, 2296, 2886, 4670],
                     index=['서울', '대구', '경기', '경북', '인천', '광주', '충남', '부산'])
plt.bar(covid_19.index, covid_19, label='확진자수', color='orange')
plt.legend()
plt.title('2021년 04월 19일 COVID19 확진현황')

for i, v in enumerate(covid_19.index):
  plt.text(v, covid_19[i], covid_19[i],
           fontsize = 10,
           color='blue',
           horizontalalignment='center',
           verticalalignment='bottom'
           )
  plt.show()
```
