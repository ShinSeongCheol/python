#iloc, loc를 사용하여 행/열 선택하기
iloc은 인덱스 번호로 데이터 프레임의 데이터를 선택하고, loc는 lable을 기준으로 선택한다.

다음과 같이 데이터 프레임을 생성하자.

```python
import numpy as np
import pandas as pd

df2d = pd.DataFrame(np.random.randint(50, 101, (5, 8)),
                    columns=['국어', '영어', '수학', '한국사', '사회', '과학', '체육', '예술'],
                    index=['김효림', '박민철', '정윤주', '최나은', '홍길동'])

df2d
```

#행번호로 선택하는 방법(iloc)

```python
#df2d데이터프레임의 첫 행만 확인
df2d.iloc[0]
```

```python
#df2d 데이터프레임의 처음 세 개의 행을 확인
df2d.iloc[:3]
```

```python
#df2d 데이터프레임의 3번 행부터 끝까지 확인
df2d.iloc[3:]
```

```python
#여학생 정보만 추출
femail = df2d.iloc[[0, 2, 3]]
femail
```

```python
#df2d 데이터프레임의 끝에서 세 번째 열을 확인
df2d.iloc[:, -3]
```

```python
#df2d 데이터프레임의 처음 세 개의 열을 확인
df2d.iloc[:, :3]
```

```python
#df2d 데이터프레임의 3번 열부터 끝까지
df2d.iloc[:, 3:]
```

```python
#국어, 수학, 과학 과목만 선택한 후 확인
choice = df2d.iloc[:, [0, 2, 5]]
choice
```

# lable을 이용하여 데이터 선택(loc)

```python
#lable이 '정윤주'인 행을 확인
df2d.loc['정윤주']
```

```python
#lable이 '박민철'과 '최나은'인 행을 확인
df2d.loc[['박민철', '최나은']]#반드시 이중의 대괄호 사용
```

```python
# '정윤주', 박민철'과 '최나은'의 '국어', '사회', '과학' 점수 확인
df2d.loc[['정윤주', '박민철', '최나은'], ['국어', '사회', '과학']]
```

# 행 추가

loc를 이용하여 행 추가

```python
df2d
```

```python
#dataframe.loc[행정보, 열정보]
df2d.loc['남효민'] = np.random.randint(50,101,8)
df2d
```

# 데이터프레임의 원소를 이용하여 계산하기


```python
#과목별 합계와 평균을 계산하여 출력하기
df2d.loc['합계', :] = df2d.iloc[:6].sum()
df2d.loc['평균', :] = df2d.iloc[:6].mean()
df2d
```

```python
df2d.loc[:, '총점'] = df2d.iloc[:, :8].sum(axis=1)
df2d.loc[:, '합계'] = df2d.iloc[:, :8].mean(axis=1)
df2d
```

**조건에 맞는 행 추출하기**

```python
new_df2d = df2d.iloc[:5]
high = new_df2d.loc[new_df2d['합계'] >= 75]
high
```

**조건에 맞는 행과 열 추출하기**

```python
high2 = df2d.loc[df2d['합계'] > 75, ['국어', '수학', '과학']]
high2
```
