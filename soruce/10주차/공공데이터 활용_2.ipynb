```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
tempData = pd.read_excel('/gdrive/My Drive/소프트웨어 이해와 활용/andongTemp.xlsx')
tempData
```

```python
#'지점' 컬럼 삭제하고 '날짜' 컬럼을 인덱스로 설정
tempData.drop('지점', axis=1, inplace=True)
tempData.set_index('날짜', inplace=True)
tempData
```

```python
#1973년 1월 1일 이후 안동의 최고기온을 상자 그래프로 표시
##상자 그래프(boxplot)를 그래서 기상 관측 이래 안동의 하루 중 최고 기온의 범위 파악
highTemp = np.array(tempData.loc[:, '최저기온(℃)'])
plt.boxplot(highTemp)
plt.show()
```

```python
#1973년 이후 안동의 1월과 8월의 최고 기온을 히스토 그램으로 그리기
jan = []
aug = []
for i in range(len(tempData)):
  month = str(tempData.index[i]).split('-')[1]
  if month == '01' :
    jan.append(tempData.iloc[i][-1])
  elif month == '08' :
    aug.append(tempData.iloc[i][-1])
plt.hist(jan, bins=50, color='b', label='Jan')
plt.hist(aug, bins=50, color='r', label='aug')
plt.legend()
plt.show()
```

```python
#1973년 이후 안동의 1월과 8월의 최고 기온을 상자 그래프로 표시
jan = []
aug = []
for i in range(len(tempData)):
  month = str(tempData.index[i]).split('-')[1]
  if month == '01' :
    jan.append(tempData.iloc[i][-1])
  elif month == '08' :
    aug.append(tempData.iloc[i][-1])
plt.xticks([1], ['high'])
plt.boxplot(jan)
plt.boxplot(aug)

plt.show()
```

```python
#상자 그래프의 위, 아래에 그려진 동그라미는 이상치를 표현한것으로 다른 수치에 비하여 너무 크거나 작은 값들을 자동으로 나타낸것
jan = []
aug = []
for i in range(len(tempData)):
  month = str(tempData.index[i]).split('-')[1]
  if month == '01' :
    jan.append(tempData.iloc[i][-1])
  elif month == '08' :
    aug.append(tempData.iloc[i][-1])
plt.xticks([1, 2], ['jan', 'aug'])
plt.boxplot([jan, aug])

plt.show()
```

```python
#아래의 코드를 다시 실행
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
tempData = pd.read_excel('/gdrive/My Drive/소프트웨어 이해와 활용/andongTemp.xlsx')
#'지점' 컬럼 삭제하고 '날짜' 컬럼을 인덱스로 설정
tempData.drop('지점', axis=1, inplace=True)
tempData.set_index('날짜', inplace=True)
```

```python
#마이너스 부호 깨짐 현상 없애기
plt.rcParams['axes.unicode_minus'] = False
jan = []
aug = []

for i in range(len(tempData)):
  month = str(tempData.index[i]).split('-')[1]
  if month == '01' :
    jan.append(float(tempData.iloc[i][-1]))
  elif month == '08' :
    aug.append(float(tempData.iloc[i][-1]))
plt.title('안동 지역 1월과 8월의 최고 기온 분석')
plt.xticks([], ['1월', '8월'])
plt.boxplot([jan, aug])

plt.show()
```

```python
#안동 지역 월별 최고 기온 분석 - 상자 그래프
plt.rcParams['axes.unicode_minus'] = False
month = [[], [], [], [], [], [], [], [] ,[] ,[] ,[] ,[]]
for i in range(len(tempData)):
  m = int(str(tempData.index[i]).split('-')[1])-1
  month[m].append(float(tempData.iloc[i][-1]))

plt.rcParams["figure.figsize"] = (12,6)
plt.title('안동 지역 최고 기온 분석')
plt.boxplot(month)

plt.show()
```

```python
#안동 지역 월별 최고 기온 분석 - 상자 그래프
plt.rcParams['axes.unicode_minus'] = False
month = [[], [], [], [], [], [], [], [] ,[] ,[] ,[] ,[]]
for i in range(len(tempData)):
  m = int(str(tempData.index[i]).split('-')[1])-1
  month[m].append(float(tempData.iloc[i][-1]))

plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (12,6)
plt.title('안동 지역 최고 기온 분석')
plt.boxplot(month)

plt.show()
```
