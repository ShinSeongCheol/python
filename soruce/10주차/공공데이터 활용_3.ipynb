```python
#지금까지의 코드를 하나의 셀에 정리하기
import numpy as np
import pandas as pd
#csv파일 읽기
popuData = pd.read_excel('/gdrive/My Drive/소프트웨어 이해와 활용/2021경상북도 인구통계 월간.xlsx')
#데이터 정제하기
popuData.drop(index=[0,2,3], inplace=True)
popuData.drop(['행정기관코드', '총 인구수', '연령구간인구수'], axis=1, inplace=True) #불필요한 열 삭제
#'행정기관'을 인덱스로 지정하고 간단하게 수정
cityName = []
for c in popuData['행정기관']:
  cityName.append(c[5:8])
del popuData['행정기관']
popuData.index = cityName
#popuData 확인
popuData
```

```python
#포항과 안동의 연령별 인구를 가로 막대그래프(항아리형 그래프)로 표시하여 비교하기
import matplotlib.pyplot as plt
subPopu = pd.DataFrame(popuData.loc[['포항시', '안동시']])
#subPopu = pd.DataFrame(popuData.iloc[[0, 3]])
subPopu = subPopu.T

subPopu
```

```python
subPopu.columns
```

```python
pohang, andong = subPopu['포항시'], subPopu['안동시']
for i in range(len(subPopu['포항시'])):
  if type(pohang[i]) == type('abc'):
    pohang[i] = pohang[i].replace(',', '')
    pohang[i] = pd.to_numeric(pohang[i])

for i in range(len(subPopu['안동시'])):
  if type(andong[i]) == type('abc'):
    andong[i] = andong[i].replace(',', '')
    andong[i] = pd.to_numeric(andong[i])

andong
```

```python
import matplotlib.pyplot as plt
plt.barh(range(101), pohang)
plt.barh(range(101), andong)
plt.show()
```

```python
import matplotlib.pyplot as plt
pohang *= -1
plt.title('포항과 안동의 인구 비교')
plt.barh(range(101), pohang, label='포항')
plt.barh(range(101), andong, label='안동')
plt.legend()
plt.rcParams["figure.figsize"] = (12, 6)
plt.show()
```
